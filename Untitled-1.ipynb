# Trazado básico y personalización de la estética
## Trazado básico

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
population = pd.read_csv('Population.csv')
print(population.shape)
population
```

```python
mexico = population.loc[population['Country Name'] == 'Mexico']
mexico = mexico.drop(columns=['Country Name','Country Code'])
mexico
```

Las funciones **plot** grafican por defecto los índices de las filas en el eje **x**, por lo tanto, se necesita transponer.

```python
mexico = mexico.T
mexico
```

```python
# Matplotlib
plt.plot(mexico.index,mexico.values)
```

```python
# Pandas
mexico.plot()
```

```python
# Seaborn
sns.lineplot(x=mexico.index,y=mexico[152])
```

## Personalización de la estética

```python
plt.plot(mexico.index,mexico.values,color='r',linestyle='--',marker='o')
```

Lo anterior se puede especificar también a través de una cadena de formato ('[marker][line][color]') que es una abreviatura para configurar rápidamente las propiedades básicas de la siguiente línea.

```python
plt.plot(mexico.index,mexico.values,'r--o')
```

```python
# Pandas
mexico.plot(title='Population of Mexico',legend = False,
            xlabel='Year',ylabel='Population',rot=45,fontsize=16)
```

Donde:
- **rot**: Indica la rotación en las marcas de los ejes en grados.
- **fontsize**: Es el tamaño de la letra en marcas de los ejes.

## Personalización a través de Matplotlib
Como cualquiero gráfico creado por Pandas y Seaborn es un objeto de Matplotlib y sus métodos plot son únicamente en un envolorio de **plt.plot()**, es posible la personalización de sus propiedades con las siguientes opciones:
- **kwargs**: permite pasar argumentos de longitud variable asociados a un nombre.
- Con las funciones de la interfaz pyplot

La interfaz pyplot provee muchas funciones para la configuración de los objetos de Matplotlib.
- **legend**: Colocar una leyenda en la figura
- **title**: Incluye el título
- **xlabel/ylabel**: Etiquetar los ejes
- **xtics/yticks**: Configurar las marcas de los ejes
- **tick_params:** cambia la apariencia de las marcas de los ejes, sus etiquetas y las líneas de la cadrícula. 

```python
sns.lineplot(x=mexico.index,y=mexico[152])
plt.title('Population of Mexico')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45,fontsize=16)
plt.show()
```

Para guardar los gráficas se usa **savefig** en el módulo pyplot de Matplotlib.
