````python
metadata = pd.read_csv('Metadata.csv')
metadata.head()
````

El dataframe **population** igual contiene la columna country code, por lo cual se puede hacer un **merge**.

````python
merged_df = pd.merge(left=metadata,right=population,left_on='Country Code',
                     right_on='Country Code')
merged_df
````

El obtejitvo será sacar el total de población por región para el año 2019.

````python
region_population = merged_df.groupby(['Region']).sum()[['2019']]
region_population
````

````python
region_population.plot(rot=90,figsize=(10,8),legend=False,
                       linestyle='-',linewidth=3,marker='s',
                       markersize=20)
plt.title('Population by region 2019',fontsize=20,
          horizontalalignment='left')
plt.xlabel('Region',fontsize=20)
````

- La API **plot()** de Pandas permite crear de forma rápida y sencilla cualquier gráfico a partir de un *dataframe*.

## Tipos de gráficos


### Gráfico de línea
- Matplotlib:
```plt.plot(df.index,'column,data=df)```
- Pandas:
```df[column].plot.line()```
- Seaborn:
```sns.relplot(kind='line',x=df.index,y='column,data=df)```

### Gráfico de barras
Para ejemplificar su uso (total,promedio u otra medida) se grafica se sigioente dataframe con la población en el 2019 de seis países.

````python
population_2019 = population.loc[population['Country Name'].isin(
    ['Mexico','Germany','Canada','Japan','United States','Nigeria']),
    ['Country Name','2019']]
````

````python
population_2019 = population_2019.set_index('Country Name')
population_2019
````

````python
population_2019.index
````

````python
# Matplotlib
plt.bar(population_2019.index,'2019',data=population_2019)
````

````python
# Pandas
population_2019['2019'].plot.bar()
````

````python
# Seaborn

sns.barplot(x=population_2019.index,y='2019',data=population_2019,
            palette='pastel',hue='2019',legend=False)
````

````python
# Seaborn horizontal
sns.barplot(x=population_2019['2019'],y=population_2019.index,
            orient='h',palette='pastel',hue=population_2019['2019'],
            legend=False)
````

¿Cómo saber que plataforma facilita la graficación de cada caso? Cuando los orígenes de los datos están normalizados es conveniendo utilizar Pandas para graficar, de lo contrario, cuando la agrupación semántica de Seaborn es más beneficiosa.

````python
population_years = population.loc[population['Country Name'].isin(
    ['Mexico','Germany','Canada','Japan','United States','Nigeria']),
    ['Country Name','2017','2018','2019']]
population_years = population_years.set_index('Country Name')
population_years
````

````python
population_years.plot.bar()
````

````python
population_years.reset_index(inplace=True)
df = population_years.melt(id_vars='Country Name',
                           var_name='Year',value_name='Population')
sns.barplot(x='Country Name',y='Population',hue='Year',data=df)
````

Se pueden aplicar otras medidas de resumen a las categorías, por ejemplo:
- Desviación Estándar

````python
# Matplotlib
df_std = df.groupby(['Country Name']).std()
plt.bar(df_std.index,df_std['Population'])
````

````python
# Pandas
df_std.plot.bar()
````

````python
# Seaborn
sns.barplot(x='Country Name', y='Population',data=df,estimator=np.std,errorbar=None)
````
