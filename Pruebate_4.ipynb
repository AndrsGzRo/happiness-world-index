1. Lee en dos dataframes: **gpd** y **unemployment**, respectivamente almacenan por país el PIB por persona empleado y la tasa de desempleo desde 1991.

El PIB por persona empleado es el producto interno bruto dividido por el total de personas empleadas en la economía.

El desempleo se refiere a la proporción de la fuerza laboral que no tiene trabajo, pero que está disponible y en busca de empleo.

```python
# Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataframes

gdp = pd.read_csv('GDP.csv')
unemployment = pd.read_csv('Unemployment.csv')
```

2. A partir de los dataframes anteriores, obtén uno (**info2019**) en donde únicamente almecenes la información de ambos indicadores en el año 2019.

```python
gdp_2019 = gdp[['Country Code','2019']]
gdp_2019
```

```python
unemployment_2019 = unemployment[['Country Code','2019']]
unemployment_2019
```

```python
info2019 = gdp_2019.merge(unemployment_2019,on='Country Code')
info2019.columns = ['Country Code', 'GDP','Unemployment']
info2019
```

3. Grafica el dataframe anterior, utilizando la plataforma Matplotlib, para visualizar la distribución de estos índices a través de histogrames en ejes duales.

```python
# Matplotlib
fig,ax1 = plt.subplots()
ax1.hist(x='GDP',data=info2019,label='GDP',alpha=0.5)
ax1.set_xlabel('GDP')
ax1.set_ylabel('Frequency')

ax2 = ax1.twiny()
ax2.hist(x='Unemployment',data=info2019,label='Unemployment',alpha=0.5,
         color='darkseagreen')
ax2.set_xlabel('Unemployment')

plt.title('World 2019')
fig.legend(bbox_to_anchor=(0.86,0.87))
plt.show()
sns.set_palette('Spectral')
```

4. Almacena en un dataframe (**continents**), el contenido del archivo **Continents.csv**, que indica a qué continente y región pertence a cada país, y combínalo con el dataframe **info2019**, uslando la columna **Country Code**, para obtener el dataframe **continents_2019**.


```python
continents = pd.read_csv('Continents.csv')
continents
```

```python
continents_2019 = info2019.merge(continents,on='Country Code')
continents_2019
```

5. Del dataframe anterior, obtén los promedios por continente de ambos índices. Almacena el resultado en **continents_group**.

```python
continents_group = continents_2019.groupby(['Continent'])[['GDP','Unemployment']].mean()
continents_group
```

6. Crea un matriz en la que vayas añadiendo las siguientes gráficas:


```python
sns.set_palette('pastel')
fig = plt.figure(figsize=(17,5))

ax1 = fig.add_subplot(131)
info2019.plot(kind='scatter',x='GDP',y='Unemployment',logx=True,ax=ax1)
ax1.set_title('World 2019')

ax2 = fig.add_subplot(132)
sns.boxplot(x='Continent',y='Unemployment',data=continents_2019,ax=ax2,
            hue='Continent')
ax2.set_title('Unemployment by Continent')

ax3 = fig.add_subplot(133)
continents_group.plot(kind='pie',y='GDP',autopct='%1.1f%%',ax=ax3)
ax3.set_title('GDP by Continent')

plt.show()
```

7. Almacena la información de México en un dataframe donde se almacene el PBI y el índice de desempleo para visualiar la evolución a través de los años.

```python
mexico_gdp = gdp[gdp['Country Code'] == 'MEX']
mexico_gdp.set_index('Country Code',inplace=True)
mexico_gdp = mexico_gdp.drop('Country Name',axis=1)
mexico_gdp = mexico_gdp.T
mexico_gdp.index.name = 'Year'
mexico_gdp.columns.name = ' '
mexico_gdp.columns = ['GDP']
mexico_gdp.head()
```

```python
mexico_unemployment = unemployment[unemployment['Country Code'] == 'MEX']
mexico_unemployment.set_index('Country Code',inplace=True)
mexico_unemployment = mexico_unemployment.drop('Country Name',axis=1)
mexico_unemployment = mexico_unemployment.T
mexico_unemployment.index.name = 'Year'
mexico_unemployment.columns.name = ''
mexico_unemployment.columns = ['Unemployment']
mexico_unemployment.head()
```

```python
mexico = mexico_gdp.join(mexico_unemployment)
mexico.head()
```

```python
sns.set_palette('pastel')

ax1 = sns.lineplot(x='Year',y='GDP',data=mexico,label='GDP')
ax1.set_ylabel('GDP')
ax1.set_xticks(np.arange(0,len(mexico)+1,4))

ax2 = ax1.twinx()
sns.lineplot(x='Year',y='Unemployment',data=mexico,label='Unemployment',
             color='lightgreen')
```
