1. Lee el dataframe **fertility** que almacena, por país, las tasas de fertilidad desde 1960.

```python
# Importando librerías
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leyendo el archivo fertility
fertility = pd.read_csv('Fertility.csv')
fertility.head()
```

2. A partir del dataframe **fertility** obtén otro, **mexico**, donde únicamente se almacene la información de México. Haz que el campo **Country Name** quede como índice.


```python
mexico = fertility.loc[fertility['Country Name'] == 'Mexico']
mexico.set_index('Country Name',inplace=True)
mexico
```

3. Elimina el campo **Country Code** e intercambia las filas por columnas

```python
mexico = mexico.drop(['Country Code'],axis=1)
mexico = mexico.T
mexico.head()
```

4. Grafica el dataframe anterior utilizando la plataforma Pandas y personaliza sus propiedades para que sea como en la siguiente imagen:


```python
mexico.plot(title='Fertility in Mexico', xlabel='Year', 
            ylabel='Fertility Rate', linestyle=':')
```

5. Realizar nuevamente el proceso anterior, pero manteniendo la información de México y Canadá.

```python
mexico_canada = fertility[(fertility['Country Name'] == 'Mexico')
                          | (fertility['Country Name'] == 'Canada')]
mexico_canada.set_index('Country Name',inplace=True)
mexico_canada
```

```python
mexico_canada = mexico_canada.drop(['Country Code'],axis=1)
mexico_canada = mexico_canada.T
mexico_canada.head()
```

```python
mexico_canada.plot(title='Fertility Rates in Mexico and Canada',
                   xlabel='Year', ylabel='Fertility Rate'
                   ,linestyle=':')
```

6. Grafica la información de 5 países (México, Estados Unidos, Somalia, España y Japón) únicamente del último año reportado. En el eje **x** deben mostrarse los nombres de los países con alineación vertical.


```python
cinco = fertility.loc[fertility['Country Name'].isin(['Mexico',
                                                      'United States',
                                                      'Somalia','Japan','Spain']),
                                                      ['Country Name','2018']]
cinco.set_index('Country Name', inplace=True)
cinco.plot(title='Fertility Rate 2018',linestyle=':',legend=False)
```
