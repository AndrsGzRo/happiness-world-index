```python
# Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Dataframes
adolescent_fertility = pd.read_csv('AdolescentFertility.csv')
mortality = pd.read_csv('Mortality.csv')
```

2. A partir de los dataframes anteriores, obtén uno donde únicamente almacene la información de México.

```python
# Adolescent Mexico
adolescent_mex = adolescent_fertility[adolescent_fertility['Country Name'] == 'Mexico']
adolescent_mex.set_index('Country Name',inplace=True)
adolescent_mex = adolescent_mex.drop(['Country Code'],axis=1)
adolescent_mex = adolescent_mex.T
adolescent_mex.index.name = 'Year'
adolescent_mex.columns.name = ''
adolescent_mex.columns = ['Adolescent Fertility']
adolescent_mex.head()
```

```python
# Mortality
mortality_mex = mortality[mortality['Country Name'] == 'Mexico']
mortality_mex.set_index('Country Name',inplace=True)
mortality_mex = mortality_mex.drop(['Country Code'],axis=1)
mortality_mex = mortality_mex.T
mortality_mex.index.name = 'Year'
mortality_mex.columns.name = ''
mortality_mex.columns = ['Mortality']
mortality_mex.head()
```

```python
# Dataframe unido
mexico = adolescent_mex.join(mortality_mex)
mexico.head()
```

3. Grafica el dataframe anterior, utilizando la plataforma Pandas, para conocer la evolución de estos índices a través de los años. Personaliza las propiedades del gráfico.

```python
# Graficano
mexico.plot()
plt.title('Mexico')
plt.ylabel('Rate x 1000')
sns.set_palette('husl')
sns.set_style('whitegrid')
```

```python
# Adolescents Fertility in North America
adolescent_na = adolescent_fertility.loc[
    adolescent_fertility['Country Name'].isin(['Mexico',
                                               'United States',
                                               'Canada']),
                                               ['Country Name','2019']]
adolescent_na.set_index('Country Name',inplace=True)
adolescent_na.columns.name = ''
adolescent_na.columns = ['Adolescent Fertility']
adolescent_na
```

```python
# Mortality in kids in North America
mortality_na = mortality.loc[
    mortality['Country Name'].isin(['Mexico','United States',
                                    'Canada']),['Country Name','2019']]
mortality_na.set_index('Country Name',inplace=True)
mortality_na.columns.name = ''
mortality_na.columns = ['Mortality']
mortality_na
```

```python
# North America
north_america = adolescent_na.join(mortality_na)
north_america
```

5. Visualiza el dataframe anterior en un gráfico de barras, ocupando Pandas.

```python
north_america.plot(kind='bar')
plt.title('North America 2019')
plt.ylabel('Rate x 1000')
sns.set_palette('husl')
```

6. Para observar la distribución de ambos indicadores en el mundo, para el 2019, construye un gráfico de histogramas independientes utilizando matplotlib. Utiliza 5 bins y una transparencia de 0.5

```python
plt.hist('2019',data=adolescent_fertility,
         label='Adolescent Fertility',bins=5,alpha=0.5)
plt.hist('2019',
         label = 'Mortality',data=mortality,bins = 5, alpha=0.5)
plt.title('World 2019')
plt.legend()
```

7. ¿Cómo han cambiado estos índices en la última decada? Compáraldos con boxplots en los años 2009 y 2019.

```python
sns.set_palette('husl')
adolescent_fertility[['2009','2019']].plot.box(
    title = 'World Adolescent Fertility',ylabel='Rate x 1000')
```
